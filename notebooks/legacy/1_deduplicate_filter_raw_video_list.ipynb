```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw_data_video_list.csv')
```

```python
df.head()
```

```python
df['title'] = df['title'].str.strip().str.replace(r'\s+', ' ')
df['status'] = df['status'].replace("Not processed", "not_processed")
```

```python
df.yt_channel.unique()
```

```python
def convert_to_minutes(time_str):
    """
    Convert a time string in HH:MM:SS or MM:SS format to total minutes.

    Parameters:
    - time_str (str): The time string in HH:MM:SS or MM:SS format.

    Returns:
    - int: The total time in minutes.

    Raises:
    - ValueError: If the time_str format is not HH:MM:SS or MM:SS.
    """
    parts = time_str.split(':')
    if len(parts) == 3:
        hours, minutes, seconds = parts
        total_minutes = int(hours) * 60 + int(minutes) + int(seconds) / 60
    elif len(parts) == 2:
        minutes, seconds = parts
        total_minutes = int(minutes) + int(seconds) / 60
    else:
        raise ValueError(f"Invalid time format: {time_str}")
    return int(total_minutes)
```

```python
df['duration_in_mins'] = df['duration'].apply(convert_to_minutes)
```

```python
output = df[(df.duration_in_mins >= 15) & (df.duration_in_mins <= 100) & (df.yt_channel == "@ElPelucaMilei")]
```
